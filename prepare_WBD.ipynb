```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
huc12 = gpd.read_file('./data/test_data/NHD_H_0108_HU4_GDB.gdb', layer = 'WBDHU12')
```

```python
wbd_buff = gpd.read_file('./data/test_data/test_wbd_buff.shp')
```

```python
wbd = huc12[huc12['HUC12'].isin(wbd_buff.HUC12)].copy()
```

```python
def getHUC8(huc12):
    return huc12[0:8]
```

```python
wbd['HUC8'] = wbd.HUC12.map(getHUC8)
```

```python
wbd.plot('HUC8')
```

```python
wbd = wbd.to_crs(epsg=5070)
```

```python
wbd.to_file('./data/test_data/test_WBD.shp')
```
